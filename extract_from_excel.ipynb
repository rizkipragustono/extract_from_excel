# Extract Name and Contact Number from Excel

```python
import pandas as pd
```

```python
df = pd.read_excel('03-12.xlsx',header=None,skiprows=1)
# Define the fields in the order they appear
fields = ["NAME", "ADRESS", "CITY", "PROVINCE", "POSTAL CODE", "COUNTRY", "TELEPHONE"]

# Split the data into chunks of the length of fields
records = [df.iloc[i:i+len(fields), 1].values for i in range(0, len(df), len(fields))]

# Create a new DataFrame
new_df = pd.DataFrame(records, columns=fields)
df = new_df.dropna(how='all')  # Drop rows where all elements are NaN
df
```

```python
df["TELEPHONE"] = df["TELEPHONE"].astype(str)
new_df = pd.DataFrame(columns=["Name", "Phone Number"])
country_codes = {"italy": "39", "france": "33", "spain": "34"}

for index, row in df.iterrows():
    if row["COUNTRY"] == "italy":
        if not row["TELEPHONE"].startswith(country_codes["italy"]):
            row["TELEPHONE"] = country_codes["italy"] + row["TELEPHONE"]
        new_df.loc[len(new_df)] = [row["NAME"], row["TELEPHONE"].replace("-", "")]
    elif row["COUNTRY"] == "france":
        if not row["TELEPHONE"].startswith(country_codes["france"]):
            row["TELEPHONE"] = country_codes["france"] + row["TELEPHONE"]
        new_df.loc[len(new_df)] = [row["NAME"], row["TELEPHONE"].replace("-", "")]
    elif row["COUNTRY"] == "spain":
        if not row["TELEPHONE"].startswith(country_codes["spain"]):
            row["TELEPHONE"] = country_codes["spain"] + row["TELEPHONE"]
        new_df.loc[len(new_df)] = [row["NAME"], row["TELEPHONE"].replace("-", "")]
new_df
        
```

```python
new_df.to_excel("final_output.xlsx", index=False)
```
